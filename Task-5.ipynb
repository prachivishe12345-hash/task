#Question no. 1

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
years = ['2003', '2007', '2011', '2015', '2019']
sachin_runs = [1141, 1425, 756, 0, 0]
kohli_runs = [0, 0, 1381, 1292, 1377]
sehwag_runs = [1319, 1212, 1000, 0, 0]
x = np.arange(len(years))
width = 0.25  

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(x - width, sachin_runs, width, label='Sachin')
plt.bar(x, kohli_runs, width, label='Kohli')
plt.bar(x + width, sehwag_runs, width, label='Sehwag')

plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs Comparison over Selected Years')
plt.xticks(x, years)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Sample Data
players = ('sachin','kholi', 'shewag')
debut_5yrs_runs = [3262, 3690, 2950]

#Plotting
plt.figure(figsize = (8,4))
plt.bar(players, debut_5yrs_runs, color=['blue', 'orange', 'green'])
plt.xlabel('Total Runs')
plt.title('Total Runs in first 5 Years of Career')
plt.tight_layout()
plt.show()
```

#Question no. 2

```python
#Pie Chart

#Sample Data
labels = ('Andriod', 'Ios', "Others")
market_share = (72, 27, 2)

#Plotting
plt.figure(figsize=(8,8))
plt.pie(market_share, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Mobile OS Market Share')
plt.show()
```

```python
#Plotting
plt.figure(figsize=(8,4))
plt.barh(labels, market_share, color=('Green', 'Gray', 'Blue'))
plt.xlabel('Market_share %')
plt.title('Mobile OS Market Share')
plt.tight_layout()
plt.show()
```

# Which one is easier to understand ?

conclusion:

-For minimum categories, the pie chart is easier to understand

-For precise comparison, especially with small values (like "iOS"),
 the horizontal bar chart is clearly more effective.
